```python
temp = []

temp_values = land_ocean_data['Temp']
for i in range(len(temp_values)-1): #creating the change in global temps compared to 1950 by using an empty list and iterating over the indexs of the Temp column in the dataset
    change_temp = abs(temp_values[i]) + temp_values[0]
    temp.append(change_temp)


fig1_2 = px.line(land_ocean_data, x=temp_values, y='Temp', title='Average of Land and Ocean Temperature Data Per Month from 1950 Globally')

fig1_2.update_layout(
    xaxis_title='Year',
    yaxis_title='Celcius',
    legend_title='Legend',
)

fig1_2.show()
```
